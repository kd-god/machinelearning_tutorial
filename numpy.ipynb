```python
import numpy as np
list1 = [1,2,3]
print('list type', type(list1))
array1 = np.array(list1)
print(array1)
print('array type', type(array1))
print('array type', array1.shape)

array2 = np.array([[1,2,3],
                   [4,5,6]])
print(array2)
print('array type', type(array2))
print('array shape', array2.shape)

array3 = np.array([[1,2,3]])
print(array3)
print('array type', type(array3))
print('array shape', array3.shape)

print('array1 : {0} 차원 , array 2 : {1}차원, array3 : {2}차원'.format(array1.ndim, array2.ndim, array3.ndim))
```

```python
list2 = [1,2,'test']
array4 = np.array(list2)
print(array4, array4.dtype)

list3 = [1.0, 2.0, 3.0]
array2 = np.array(list3).astype('int32')
print(array2, array2.dtype)
```

```python
array1 = np.arange(10)
print(array1.reshape(-1,5))
```

```python
# argsort
# 2022.07.25
```

```python
org_array = np.array([3,1,9,5])
sort_indices = np.argsort(org_array)
print(sort_indices[::-1])
```

```python
name_array = np.array(['John','Mike','Sarah','Kate','Samuel'])
score_array = np.array([78,95,84,98,85])

score_array_asc = np.argsort(score_array)
print(name_array[score_array_asc])
```
